# Pandas 연습문제

- 문제 1: 기상 관측 이래, 서울의 최고 기온이 가장 높았던 날은 언제였고, 몇도인가요?
- 문제 2: 서울 지역에서, 역사적으로 일교차가 가장 큰 날짜는 몇년 몇월 몇일 인가요?
- 문제 3: 서울 지역에서, 1년 중 평균적으로 일교차가 가장 큰 날짜는 몇월 몇일 인가요?
- 문제 4: 가장 덥다고 알려진 대구보다 서울이 더 더운날이 가장 많은 연도는 언제인가요?

```python
import numpy as np
import pandas as pd
print('modules imported!')

seoul = pd.read_csv('./data/seoul.csv')
daegu = pd.read_csv('./data/daegu.csv')
print('CSVs loaded!')

display(seoul.head())
display(daegu.head())
```

```python
# 문제 1: 기상 관측 이래, 서울의 최고 기온이 가장 높았던 날은 언제였고, 몇도인가요?

sorted = seoul.sort_values(by='max', ascending=False)
print(sorted.iloc[0, [0, 4]])
```

```python
# 문제 2: 서울 지역에서, 역사적으로 일교차가 가장 큰 날짜는 몇년 몇월 몇일 인가요?

seoul['differences'] = seoul['max'] - seoul['min']
sorted = seoul.sort_values(by='differences', ascending=False)
print(sorted.iloc[0, 0])
```

```python
# 문제 3: 서울 지역에서, 1년 중 평균적으로 일교차가 가장 큰 날짜는 몇월 몇일 인가요?

seoul['year'] = seoul['date'].apply(lambda x: x[:4])
group_by_year = seoul.groupby(['year'])

for key, year in group_by_year:
    sorted = year.sort_values(by='differences', ascending=False)
    print(sorted.iloc[0, 0])
```

```python
# 문제 4: 가장 덥다고 알려진 대구보다 서울이 더 더운날이 가장 많은 연도는 언제인가요?

merged = pd.merge(seoul, daegu, on='date', how='inner')
display(merged)
```
